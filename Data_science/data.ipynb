```python
obesityTypeDataSet = obesityDataSet.loc[(obesityDataSet['NObeyesdad'] != 'Normal_Weight') & (obesityDataSet['NObeyesdad'] != 'Overweight_Level_I') & (obesityDataSet['NObeyesdad'] != 'Overweight_Level_I') & (obesityDataSet['NObeyesdad'] != 'Overweight_Level_II')]
display(obesityTypeDataSet)

```

```python
obesityTypeDataSet.loc[:, 'Age'] = obesityTypeDataSet['Age'].astype(int)
display(obesityTypeDataSet)
```

```python
obesityLevel_I = obesityTypeDataSet.loc[obesityTypeDataSet['NObeyesdad'] == 'Obesity_Type_I']
display(obesityLevel_I)
```

```python
mean_age_male_Obesity_I = obesityLevel_I.loc[obesityLevel_I['Gender'] == 'Male', 'Age'].mean()


mean_age_female_Obesity_I = obesityLevel_I.loc[obesityLevel_I['Gender'] == 'Female', 'Age'].mean()

print("Média de idade entre homens:", mean_age_male_Obesity_I)
print("Média de idade entre mulheres:", mean_age_female_Obesity_I)
```
